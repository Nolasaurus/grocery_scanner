```python
import cv2
from matplotlib import pyplot as plt

def display_image_inline(image_path):
    # Read the image using OpenCV
    image = cv2.imread(image_path)
    if image is not None:
        # Convert the image from BGR (OpenCV format) to RGB (Matplotlib format)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        # Display the image using Matplotlib
        plt.imshow(image_rgb)
        plt.title('Loaded Image')
        plt.axis('off')  # Hide axes for better visual
        plt.show()
    else:
        print("Failed to load image. Check the file path or format.")

# Example usage
image_path = 'static/captures/capture_20250110_163128.jpg'
display_image_inline(image_path)
```
